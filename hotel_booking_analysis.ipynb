```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
hotel_booking_cancellation = pd.read_csv("hotel_booking.csv")
```

```python
hotel_booking_cancellation.drop(['company','email','phone-number','credit_card'] , axis=1 , inplace=True) 
```

```python
hotel_booking_cancellation.info()
```

```python
hotel_booking_cancellation['country'] = hotel_booking_cancellation['country'].fillna('unknown')
```

```python
hotel_booking_cancellation['children'].median()
```

```python
hotel_booking_cancellation['children'] = hotel_booking_cancellation['children'].fillna(hotel_booking_cancellation['children'].median())
```

```python
hotel_booking_cancellation['agent'] = hotel_booking_cancellation['agent'].fillna(0)
```

```python
hotel_booking_cancellation.isnull().sum()
```

```python
hotel_booking_cancellation.head(10)
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December']

hotel_booking_cancellation['arrival_date_month'] = pd.Categorical(hotel_booking_cancellation['arrival_date_month'],categories=month_order,ordered=True)
```

```python
yearly_bookings = hotel_booking_cancellation.groupby('arrival_date_year').size()
yearly_bookings
```

```python
yearly_bookings.plot(
    kind='bar',
    figsize=(10,5)
)
plt.title('Yearly Booking Status')
plt.xlabel('Year')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
hotel_type = hotel_booking_cancellation.groupby(['arrival_date_year','hotel']).size().unstack()
hotel_type
```

```python
hotel_type.plot(
    kind='barh',
    figsize=(10,5)
)

plt.title('Bookings by Hotel Type and Year')
plt.xlabel('Number of Bookings')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
year_and_monthly_booking = hotel_booking_cancellation.groupby(['arrival_date_year','arrival_date_month'],observed=True).size()
year_and_monthly_booking
```

```python
year_and_monthly_booking.plot(
    kind='line',
    figsize=(12,6),
    marker='o'
)

plt.title('Monthly Hotel Bookings by Year')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
yearly_cancelled = hotel_booking_cancellation.groupby(['arrival_date_year','is_canceled']).size().unstack()
yearly_cancelled
```

```python
yearly_cancelled.plot(
    kind='bar',
    figsize=(10,5)
)
plt.title('Cancelled vs Non-Cancelled Bookings by Year')
plt.xlabel('Year')
plt.ylabel('Number Of Bookings')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
hotel_cancelled = hotel_booking_cancellation.groupby(['hotel','is_canceled']).size().unstack()
hotel_cancelled
```

```python
hotel_cancelled.plot(
    kind='bar',
    figsize=(8,5)
)
plt.title('Cancelled vs Non-Cancelled Bookings by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
hotel_booking_cancellation.groupby('is_canceled')['lead_time'].mean()
```

```python
hotel_booking_cancellation.groupby('is_canceled')['lead_time'].median()
```

```python
market_segment = hotel_booking_cancellation.groupby(['market_segment','is_canceled']).size().unstack()
market_segment
```

```python
market_segment.plot(
    kind='barh',
    figsize=(10,6)
)
plt.title('Cancelled vs Non-Cancelled Bookings by Market Segment')
plt.xlabel('Number of Bookings')
plt.ylabel('Market Segment')
plt.tight_layout()
plt.show()
```

```python
total_booking_market_segment = hotel_booking_cancellation.groupby('market_segment').size()
total_booking_market_segment
```

```python
cancellation_by_market_segment = hotel_booking_cancellation[hotel_booking_cancellation['is_canceled']==1].groupby('market_segment').size().sort_values(ascending=False)
cancellation_by_market_segment
```

```python
cancellation_rate_by_market_segment = (cancellation_by_market_segment / total_booking_market_segment) * 100
cancellation_rate_by_market_segment = cancellation_rate_by_market_segment.sort_values(ascending=True)
cancellation_rate_by_market_segment
```

```python
cancellation_rate_by_market_segment.plot(
    kind='barh',
    figsize=(12,6)
)
plt.title('Cancellation Rate By Market Segment')
plt.xlabel('Cancellation Rate (%)')
plt.ylabel('Market Segment')
plt.tight_layout()
plt.show()
```

```python
top_booking_country = hotel_booking_cancellation.groupby('country').size().sort_values(ascending=False).head(15)
top_booking_country = top_booking_country.sort_values(ascending=True)
top_booking_country
```

```python
top_booking_country.plot(
    kind='barh',
    figsize=(12,6)
)
plt.title('Top 15 Countries by Number of Bookings')
plt.xlabel('Number Of Bookings')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
total_booking = hotel_booking_cancellation.groupby('deposit_type').size()
total_booking
```

```python
cancelled_booking = hotel_booking_cancellation[hotel_booking_cancellation['is_canceled']==1].groupby('deposit_type').size()
cancelled_booking
```

```python
cancellation_rate = (cancelled_booking / total_booking) * 100
cancellation_rate
```

```python
cancellation_rate.plot(
    kind='bar',
    figsize=(10,5)
)
plt.title('Cancellation Rate By Deposit Type')
plt.xlabel('Deposit Type')
plt.ylabel('Cancellation Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
customer_type = hotel_booking_cancellation.groupby(['customer_type','is_canceled']).size().unstack()
customer_type
```

```python
customer_type.plot(
    kind='barh',
    figsize=(12,6)
)
plt.title('Cancelled vs Non-Cancelled Bookings by Customer Type')
plt.xlabel('Number of Bookings')
plt.ylabel('Customer Type')
plt.tight_layout()
plt.show()
```

```python
total_booking_by_custype = hotel_booking_cancellation.groupby('customer_type').size()
total_booking_by_custype
```

```python
cancellation_by_custype = hotel_booking_cancellation[hotel_booking_cancellation['is_canceled']==1].groupby('customer_type').size().sort_values(ascending=False)
cancellation_by_custype
```

```python
cancellation_per_by_custype = (cancellation_by_custype / total_booking_by_custype) * 100
cancellation_per_by_custype
```

```python
cancellation_per_by_custype.plot(
    kind='bar',
    figsize=(10,5)
)
plt.title('Cancellation Rate By Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Cancellation Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
